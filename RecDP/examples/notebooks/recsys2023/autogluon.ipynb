```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import sklearn.metrics
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 150)

data = pd.read_parquet("/mnt/DP_disk1/ht/datasets/autofe/recsys2023/recsys2023_train.parquet")
target_label = "is_installed"
data
```

```python
from autogluon.tabular import TabularPredictor
from pyrecdp.core.utils import Timer

features_train = [i for i in data.columns if i != target_label]
day_i_y = 66
train_df = data[data['f_1'] < day_i_y]
test_df = data[data['f_1'] == day_i_y]

hyperparameters = {
    'GBM': {}, 
    'RF': {},
    'XT': {},
#     'KNN': {},
    'LR': {},
}

with Timer(f"autogluon"):
    model = TabularPredictor(label='is_installed', eval_metric='roc_auc')
    predictor = model.fit(train_data=train_df, hyperparameters = hyperparameters)
```

```python
y_test = test_df[target_label]
test_df_nolab = test_df.drop(columns=[target_label])
pred_probs = predictor.predict_proba(test_df_nolab)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=pred_probs, auxiliary_metrics=True)
```

```python
from sklearn.metrics import log_loss
log_loss(y_test, pred_probs)
```
