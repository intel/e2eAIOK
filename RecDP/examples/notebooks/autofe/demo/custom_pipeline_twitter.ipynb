# Load pre-created pipeline and modify - twitter

## tasks:

1. RecDP will load pre-created pipeline json
2. user can modify pipeline

## 1. Load Datasets

```python
import pandas as pd
from utils import Timer
from pathlib import Path
import os, sys
pathlib = str(Path(os.path.abspath('')).parent.parent.parent.parent.resolve())

train_data = pd.read_parquet(f"{pathlib}/tests/data/test_twitter_recsys.parquet")
```

## 2. Load pipeline

```python
from pyrecdp.autofe import BasePipeline
# create a brand-new pipeline and import from json
new_pipeline = BasePipeline(dataset=train_data, label="reply")
new_pipeline.import_from_json(f"twitter_pipeline.json")
new_pipeline.plot()
```

## 3. execute pipeline with pandas

```python
transformed = new_pipeline.fit_transform()
transformed
```
